 What is the difference between descriptive statistics and inferential
statistics? Explain with examples
- Descriptive statistics summarize and describe the features of a dataset.
They deal with what the data shows.
- Purpose

- To present data in a meaningful way.

- To provide quick insights into patterns or characteristics

- Common Descriptive Tools

- Mean, median, mode

- Standard deviation, variance

- Minimum, maximum, range

- Tables, charts, histograms

- Inferential Statistics

Inferential statistics help you make predictions or draw conclusions about a larger population based on a sample of data.

Purpose

To infer something about the entire population.

To test hypotheses or make forecasts.


 What is sampling in statistics? Explain the differences between random
and stratified sampling
- Sampling is the process of selecting a small group (sample) from a larger group (population) to study.
Since studying an entire population is often expensive or impossible, sampling allows us to collect data efficiently and make inferences about the whole population.
- Stratified Sampling
Meaning

The population is divided into subgroups (strata) based on some characteristic, and then samples are taken from each subgroup

- How it works

- Divide population by gender, age group, class, income level, etc.

- Randomly select samples from every group proportionally.

- Features

- More accurate when the population is diverse/heterogeneous.

- Ensures every important subgroup is represented.

 Define mean, median, and mode. Explain why these measures of central
tendency are important.

The mean is the sum of all values divided by the number of values.

- Median
Definition:

The median is the middle value when the data is arranged in ascending or descending order.

If the number of observations is odd, the median is the middle value.

If the number is even, the median is the average of the two middle values.

- Mode
Definition:

The mode is the value that appears most frequently in a dataset.

Example:

Values: 2, 3, 3, 5, 7 → Mode = 3

A dataset can have:

No mode

One mode

More than one mode

Summarize large data easily
They reduce a large amount of data to a single representative value.

Identify the “typical” value
They help understand what is common or expected in the dataset.

Useful for comparing groups
For example, comparing average marks of two classes.

Basis for further analysis
Many statistical tools (standard deviation, variance, correlation, regression) use the mean or median.

Handle different types of data

Mean is useful for numerical data.

Median is useful when data has extreme values or outliers.

Mode is useful for categorical data (like most common color, brand, or choice).

 Explain skewness and kurtosis. What does a positive skew imply about the
data?

- Skewness

Definition:

- Skewness measures the asymmetry of a distribution (how data is spread around the mean).

- If the left and right sides of the graph are mirror images → Skewness = 0 (symmetrical).

- If the graph leans more to one side → data is skewed.

- Types of Skewness

- Positive Skew (Right Skew)

- Negative Skew (Left Skew)

 Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)
-


```python


import statistics as st

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = st.mean(numbers)
median = st.median(numbers)
mode = st.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

 Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
(Include your Python code and output in the code box below.)

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance
cov_matrix = np.cov(list_x, list_y, ddof=1)
cov_xy = cov_matrix[0, 1]

# Correlation coefficient
corr_xy = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```

Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
(Include your Python code and output in the code box below.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Compute quartiles and IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Plot boxplot
plt.figure(figsize=(6,4))
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

q1, q3, iqr, lower_bound, upper_bound, outliers
```
