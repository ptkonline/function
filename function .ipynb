Explain the concept of generators in Python and how they are defined.

1. What is a Generator in Python?

- A generator is a special type of iterator that yields values one at a time as you loop through it, without storing the entire sequence in memory.

-  It is memory-efficient and lazy, meaning it produces values on the fly only when needed.
2.  How are Generators Defined

Generators are created in two ways:

A. Using a generator function (with yield)

Instead of return, it uses the yield keyword.

Each time the function is called, it pauses and remembers its state until the next call.

Example:
def count_up_to(n):
    count = 1
    while count <= n:
        yield count    # yields the next value
        count += 1

# Create generator object
counter = count_up_to(5)

# Access values one by one
for num in counter:
    print(num)


 What are the advantages of using generators over regular functions?
 1.  Memory Efficiency

- Regular functions compute all values at once and store them in memory (like a list).

- Generators produce values one at a time and do not store the entire sequence in memory.
- This makes them ideal for handling large datasets or infinite sequences.
2. Lazy Evaluation

- Generators generate values only when needed, not all at once.

- This means computation happens on the fly, saving time and memory when not all results are needed.
3. Infinite Sequences

- Generators can represent infinite data streams because they yield values one at a time.

- Regular functions can’t do this — they would run out of memory.

4. Better Performance for Large Data

- Because generators yield items one by one, they often perform faster and use fewer resources when processing large inputs.

5. Cleaner and Simpler Code

- Generators are simpler than writing a full iterator class with __iter__() and __next__() methods.

- Using yield makes the code more readable and concise.

| **Advantage**             | **Explanation**                     |
| ------------------------- | ----------------------------------- |
| **Memory Efficient**      | Doesn’t store all items at once     |
| **Lazy Evaluation**       | Produces values only when needed    |
| **Handles Infinite Data** | Can generate endless sequences      |
| **Faster for Large Data** | Reduces processing overhead         |
| **Simpler Code**          | Easier than custom iterator classes |

 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.




```python
def sum_even_numbers(numbers):
    """
    This function takes a list of numbers and returns the sum of all even numbers in the list.
    """
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even_numbers(nums))
```

. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_list(numbers):
    """
    This function takes a list of integers and returns a new list containing the squares of each number.
    """
    squared = [num ** 2 for num in numbers]
    return squared

nums = [1, 2, 3, 4, 5]
print("Squared list:", square_list(nums))
```

Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    """
    This function checks if a number n is prime.
    Returns True if prime, False otherwise.
    """
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True


for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.index = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.n_terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        elif self.index == 1:
            self.index += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.index += 1
            return self.b


fib = Fibonacci(10)
for num in fib:
    print(num)
```

Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    """
    Generator function that yields powers of 2 from 0 to max_exponent.
    """
    for exp in range(max_exponent + 1):
        yield 2 ** exp


for value in powers_of_two(5):
    print(value)
```

Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line and yields each line as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
```

. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32


celsius_temps = [0, 20, 37, 100]


fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python

def is_not_vowel(char):
    return char.lower() not in 'aeiou'


text = "Hello, World!"


filtered_text = ''.join(filter(is_not_vowel, text))

print("Original string:", text)
print("String without vowels:", filtered_text)
```
